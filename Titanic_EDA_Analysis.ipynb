```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)

print(data.columns)
```

Survival Count Graph

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='Survived', data=data)
plt.title("Survival Cout")
plt.show()
data['Survived'].value_counts()
```

549 passengers died                                               
342 passengers survived,                                           More passengers did not survive than those who survived

Survival by Gender

```python
sns.countplot(x='Sex', hue='Survived', data=data)
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of passengers")
plt.show()
```

This bar chart shows survival vs death based on gender.                          The blue bar shows death count ,and yellow bar shows survival count

Female passengers survived more than male passengers

Survival by passenger class

```python
sns.countplot(x='Pclass', hue='Survived', data=data)
plt.title("survival by passenger class")
plt.show()
```

passengers in 1st class had a higher survival rate than those in 2nd and 3rd class.

Age Distribution

```python
sns.histplot(data['Age'], bins=30)
plt.title("Age Distribution of passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Most passengers were between 20-40 years old.

Exploratory Data Analysis (EDA) was performed to understand the dataset patterns.various graps were created to analyze survival distribution, gender impact, passenger class influence, and age distribution. the analysis showed that female passengers and first-class passengers had higher survival rates.
